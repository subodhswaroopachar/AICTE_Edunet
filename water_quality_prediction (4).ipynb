```python
pollutants = ['O2', 'NO3', 'NO2', 'SO4',
       'PO4', 'CL']
```

```python
# drop the missing values - dropna()
df = df.dropna(subset=pollutants)
df.head()
```

```python
import matplotlib.pyplot as plt

# Create subplots for each pollutant
fig, axes = plt.subplots(2, 3, figsize=(14, 8))
axes = axes.flatten()

for i, pollutant in enumerate(pollutants):
    ax = axes[i]
    df[pollutant].hist(ax=ax, bins=30, color='skyblue', edgecolor='black')
    ax.set_title(f'Distribution of {pollutant}')
    ax.set_xlabel(f'{pollutant} Concentration (mg/L)')
    ax.set_ylabel('Number of Samples')
    ax.grid(True)

plt.suptitle("Distribution of Water Pollutants", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])  # leave space for suptitle
plt.show()
```

```python
df.isnull().sum()
```

```python
# Feature and target selection - Feature - independent variable and Target dependent variable
X = df[['id', 'year']]
y = df[pollutants]
```

```python
# Encoding - onehotencoder - 22 stations - 1 - 1
X_encoded = pd.get_dummies(X, columns=['id'], drop_first=True)
```

```python
# Train, Test and Split
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)
```

```python
# Train the model
model = MultiOutputRegressor(RandomForestRegressor(n_estimators=100, random_state=42))
model.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt

# Create a DataFrame of feature importances
importances_df = pd.DataFrame({'Feature': X_train.columns})
for i, p in enumerate(pollutants):
    rf = model.estimators_[i]
    importances_df[p] = rf.feature_importances_

# Set up 2x3 grid for 6 pollutants
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, pollutant in enumerate(pollutants):
    ax = axes[i]
    importances_df.set_index('Feature')[pollutant].sort_values().plot(
        kind='barh', ax=ax, color='skyblue', edgecolor='black'
    )
    ax.set_title(f"Feature Importance for {pollutant}")
    ax.set_xlabel("Importance")
    ax.set_ylabel("Feature")
    ax.grid(True)

plt.tight_layout()
plt.suptitle("Feature Importance per Pollutant", fontsize=18, y=1.02)
plt.show()
```

```python
# Evaluate model
y_pred = model.predict(X_test)
```

```python
print("Model Performance on the Test Data:")
for i, pollutant in enumerate(pollutants):
    print(f'{pollutant}:')
    print('   MSE:', mean_squared_error(y_test.iloc[:, i], y_pred[:, i]))
    print('   R2:', r2_score(y_test.iloc[:, i], y_pred[:, i]))
    print()
```

```python
station_id = '22'
year_input = 2024

input_data = pd.DataFrame({'year': [year_input], 'id': [station_id]})
input_encoded = pd.get_dummies(input_data, columns=['id'])

# Align with training feature columns
missing_cols = set(X_encoded.columns) - set(input_encoded.columns)
for col in missing_cols:
    input_encoded[col] = 0
input_encoded = input_encoded[X_encoded.columns]  # reorder columns

# Predict pollutants
predicted_pollutants = model.predict(input_encoded)[0]

print(f"\nPredicted pollutant levels for station '{station_id}' in {year_input}:")
for p, val in zip(pollutants, predicted_pollutants):
    print(f"  {p}: {val:.2f}")
```
